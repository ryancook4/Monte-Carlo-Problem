```python
# split up string to define vars
x_point = int(f_read.split(',')[0])
y_point = int(f_read.split(',')[1].split('\n')[0])
n_trials = int(f_read.split('\n')[1])

dist_origin = 2/3
dist_point = 1.1318
```

```python
# edited from...
## https://math.stackexchange.com/questions/3533713/uniform-random-distribution-on-a-unit-disk
from scipy.stats import uniform
import matplotlib.pyplot as plt
import math  
import numpy as np

# create vars to make unit disk
r = uniform.rvs(scale =dist_origin, size=n_trials)
pi = 3.14159265359
theta = uniform.rvs(scale =2*pi,size=n_trials)

# initialize empty lists
x = []
y = []
dist = []

# loop through trials
for i in range (n_trials):
    # x val of random point
    x_idx = r[i]*math.cos(theta[i])
    x.append(x_idx)
    
    # y val of random point
    y_idx = r[i]*math.sin(theta[i])
    y.append(y_idx)
    
    # distance calculation from (1,0) 
    dist_idx = np.sqrt(np.square(x_idx - x_point) + np.square(y_idx - y_point))
    dist.append(dist_idx)

# plot the unit disk
fig=plt.figure()
ax=fig.add_axes([0,0,2,3])
ax.scatter(x, y)

dist_mean = round(np.mean(dist), 4)
dist_diff = round(dist_mean - dist_point, 4)

print('The mean distance of the points from (1,0) is {0}, which is {1} away from the expected distance {2}.' 
      .format(dist_mean, dist_diff, dist_point))
```
